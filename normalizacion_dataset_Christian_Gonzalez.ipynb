```python
#Importar librerias
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
#Cargar el dataset
from google.colab import files

uploaded = files.upload()
#Leer el dataset
df = pd.read_csv('sales_data_sample.csv', encoding='latin1')
#Leer las primeras filas
df.head()
#Ver los tipos de datos
df.info()
#Seleccionar columnas numericas
columnas_numericas = df.select_dtypes(include=['int64', 'float64']).columns

print(columnas_numericas)
#Normalizar los datos con el mÃ©todo Min-Max Scaling
scaler = MinMaxScaler()

df_normalizado = df.copy()

df_normalizado[columnas_numericas] = scaler.fit_transform(df[columnas_numericas])
#Ver resultados
df_normalizado.head()

#COMPARAR LOS RESULTADOS
#Daraset Original
df[columnas_numericas].head()
#Dataset Normalizado
df_normalizado[columnas_numericas].head()
```

```python
#Guardar el dataset normalizado

df_normalizado.to_csv('sales_data_normalizado.csv', index=False)

#Descargar archivo con el dataset normalizado
files.download('sales_data_normalizado.csv')
```
